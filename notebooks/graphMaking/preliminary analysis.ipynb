```python
import geopandas as gpd
```

```python
# import SAL
sal_code = gpd.read_file("../../data/landing/SAL_data/SAL_2021_AUST_GDA2020.shp")

# Drop non-integer SAL_CODE21 values
gdf = sal_code[sal_code['SAL_CODE21'].apply(lambda x: str(x).isdigit())]
gdf["SAL_CODE21"] = gdf["SAL_CODE21"].astype(int)
```

```python
dfGeom = df.merge(gdf, right_on='SAL_CODE21', left_on='SAL_CODE')
```

```python
from math import radians, cos, sin, asin, sqrt

# for better visualisaiton, remove suburbs too far outside melbourne
# Compute distance to Melbourne CBD
melbourne_cbd_coords = (-37.8136, 144.9631)  # Latitude and Longitude of Melbourne CBD

# Function to calculate distance
def haversine(lat1, lon1, lat2, lon2):
    # convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    # haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Radius of earth in kilometers
    return c * r

# Apply the function to calculate distance to Melbourne CBD
dfGeom['distance_to_CBD'] = dfGeom.apply(lambda row: haversine(row['geometry'].centroid.y, row['geometry'].centroid.x, melbourne_cbd_coords[0], melbourne_cbd_coords[1]), axis=1)

# Filter out suburbs too far outside Melbourne (e.g., more than 50 km)
dfGeom = dfGeom[dfGeom['distance_to_CBD'] <= 100]
```

```python
# Filter for the year 2022
df_2022 = dfGeom[dfGeom['year'] == 2022]

# Ensure the GeoDataFrame has a geometry column for plotting
if 'geometry' in df_2022.columns:
    df_2022 = gpd.GeoDataFrame(df_2022, geometry='geometry')

# Plot the data as a choropleth map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plot the GeoDataFrame, coloring by 'A Crimes against the person' column
df_2022.plot(
    column='A Crimes against the person',  # Data column for coloring
    cmap='YlOrRd',  # Color map
    linewidth=0.8,  # Line width between polygons
    ax=ax,          # Axis to plot on
    edgecolor='0.8', # Edge color
    legend=True     # Add a legend
)

# Add a title
ax.set_title('A Crimes against the person - 2022', fontsize=15)

# Show the plot
plt.show()
```
