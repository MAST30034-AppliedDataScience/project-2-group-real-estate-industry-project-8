```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
restaurants = gpd.read_file("../../data/landing/restaurants_nodes.geojson")
shape_file = gpd.read_file("../../data/landing/SAL_data/SAL_2021_AUST_GDA2020.shp")
```

```python
# inner join
restaurants_in_SAL = gpd.sjoin(restaurants, shape_file, how='inner')
```

```python
restaurants_in_SAL
```

```python
# group by SALCODE21
restaurants_in_SAL = restaurants_in_SAL.groupby('SAL_NAME21').size().reset_index(name='count')
```

```python
restaurants_in_SAL
```

```python
# get top 10
restaurants_in_SAL = restaurants_in_SAL.nlargest(10, 'count')
restaurants_in_SAL
```
