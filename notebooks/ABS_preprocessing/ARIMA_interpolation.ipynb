```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
```

```python


# Example data loading (replace these with actual datasets)
df_2016 = pd.read_csv('../../data/curated/ABS_data_2016.csv')  # Uncomment and replace with your file path
df_2021 = pd.read_csv('../../data/curated/ABS_data_2021.csv')  # Uncomment and replace with your file path

# Sample DataFrame structure for demonstration
# Assuming the structure is something like:
# ABS_data_2016 = pd.DataFrame({'SAL': [1, 2, 3], 'Value': [100, 200, 300]})
# ABS_data_2021 = pd.DataFrame({'SAL': [1, 2, 3], 'Value': [110, 210, 310]})

```

```python
import pandas as pd

# Merge the DataFrames on 'SAL_CODE'
df = df_2016.merge(df_2021, on='SAL_CODE', suffixes=('_2016', '_2021'))

variables = df_2016.columns[1:]  # Assuming the first column is 'SAL_CODE'

# Calculate the annual rate for each variable
for col in variables:
    df[f'{col}_annual_rate'] = (df[f'{col}_2021'] - df[f'{col}_2016']) / 5

# Generate data for each year from 2016 to 2028
years = range(2016, 2029)
dfs = []
for year in years:
    temp_df = df[['SAL_CODE']].copy()
    for col in variables:
        temp_df[col] = df[f'{col}_2016'] + df[f'{col}_annual_rate'] * (year - 2016)
    temp_df['Year'] = year
    dfs.append(temp_df)

# Combine all years into a single DataFrame
result_df = pd.concat(dfs)
```

```python
# group by year and export

import os
os.makedirs('../../data/curated/ABS_extrapolated', exist_ok=True)  # Uncomment and replace with your folder path

for year in range(2016,2029):
    res = result_df[result_df['Year'] == year].drop(columns='Year')
    res.to_csv(f'../../data/curated/ABS_extrapolated/ABS_data_{year}.csv', index=False)  # Uncomment and replace with your file path
```
